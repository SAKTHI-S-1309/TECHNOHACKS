# SIMPLE CONTACT BOOK

## Problem Statement
Create a command-line contact book application that allows users to add, search, and delete contacts.



```python
contacts = {}

# Add Contact
def add_contact():
    name = input("Enter contact name: ").strip()

    if name in contacts:
        print("Contact already exists!")
    else:
        phone = input("Enter phone number: ").strip()
        contacts[name] = phone
        print("Contact added successfully!")

# Search Contact
def search_contact():
    name = input("Enter contact name to search: ").strip()

    if name in contacts:
        print(f"{name} : {contacts[name]}")
    else:
        print("Contact not found!")

# Delete Contact
def delete_contact():
    name = input("Enter contact name to delete: ").strip()

    if name in contacts:
        del contacts[name]
        print("Contact deleted successfully!")
    else:
        print("Contact not found!")

# View All Contacts
def view_contacts():
    if not contacts:
        print("No contacts available.")
    else:
        print("\n Contact List ")
        for name, phone in contacts.items():
            print(f"{name} : {phone}")

while True:
    print("\n CONTACT BOOK ")
    print("1. Add Contact")
    print("2. Search Contact")
    print("3. Delete Contact")
    print("4. View All Contacts")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_contact()

    elif choice == "2":
        search_contact()

    elif choice == "3":
        delete_contact()

    elif choice == "4":
        view_contacts()

    elif choice == "5":
        print("Exiting Contact Book")
        break

    else:
        print("Invalid choice! Please try again.")
```

## Conclusion
The Contact Book application works through a command-line interface where users can manage contact details easily. The program stores contact names and phone numbers using a Python dictionary. Users can choose different options from the menu such as adding a new contact, searching for an existing contact, deleting a contact, or viewing all saved contacts. Based on the user’s input, the corresponding function is executed, and the application displays the required result. The program continues running until the user selects the exit option.
