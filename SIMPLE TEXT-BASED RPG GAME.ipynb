# SIMPLE TEXT-BASED RPG GAME

## Problem Statement
Develop a text-based RPG (Role-Playing Game) where users can make choices and explore different outcomes.



```python
def start_game():
    print("\nWELCOME TO THE ADVENTURE GAME!")
    print("You are standing in front of a dark forest.")

    choice1 = input("Do you want to enter the forest? (yes/no): ").lower()

    if choice1 == "yes":
        forest_path()
    else:
        print("You stayed safe at home. Game Over!")

def forest_path():
    print("\nYou enter the forest and find two paths.")
    
    choice2 = input("Choose left or right: ").lower()

    if choice2 == "right":
        treasure_room()
    elif choice2 == "left":
        monster_room()
    else:
        print("Invalid choice. Game Over!")

def treasure_room():
    print("\nYou found the treasure!")
    print("YOU WON")

def monster_room():
    print("\nA wild monster appears!")
    
    choice3 = input("Fight or run?: ").lower()

    if choice3 == "fight":
        print("You fought bravely and escaped safely. Game Over!")
    elif choice3 == "run":
        print("You got caught. Game Over!")
    else:
        print("Invalid choice. Game Over!")

start_game()
```

## Conclusion
The Text-Based RPG Game is an interactive Python program where the user makes choices that affect the storyline. The game starts with a scenario and prompts the user for decisions at each stage. Based on the input, different functions are executed, leading to multiple outcomes such as winning the game, finding treasure, or facing a monster. This project demonstrates functions, conditional statements, and user interaction in Python.
