# SIMPLE REMINDER APPLICATION

## Problem Statement
Build a command-line reminder application that allows users to set reminders for specific times.



```python
import time

reminders = []

# Add Reminder
def add_reminder():
    message = input("Enter reminder message: ")
    reminder_time = input("Enter time (HH:MM:SS): ")

    reminders.append({
        "message": message,
        "time": reminder_time
    })

    print("Reminder added successfully!")

# View Reminders
def view_reminders():
    if not reminders:
        print("No reminders available.")
    else:
        print("\n REMINDERS ")

        for index, reminder in enumerate(reminders, start=1):
            print(f"{index}. {reminder['message']} at {reminder['time']}")

# Check Reminders
def check_reminders():
    print("Reminder system started")

    while True:
        current_time = time.strftime("%H:%M:%S")

        for reminder in reminders:
            if current_time == reminder["time"]:
                print(f"\nREMINDER: {reminder['message']}")

        time.sleep(1)

while True:
    print("\n REMINDER APPLICATION ")
    print("1. Add Reminder")
    print("2. View Reminders")
    print("3. Start Reminder System")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_reminder()

    elif choice == "2":
        view_reminders()

    elif choice == "3":
        check_reminders()

    elif choice == "4":
        print("Exiting Reminder Application")
        break

    else:
        print("Invalid choice!")
```

## Conclusion:
The Reminder Application works through a command-line interface where users can add and manage reminders. The program stores reminder messages and their scheduled times using a list of dictionaries. Users can add reminders by entering a message and a specific time in HH:MM:SS format. The application continuously checks the system time using the time library and triggers the reminder message when the current time matches the stored reminder time. Users can also view all saved reminders through the menu options.
