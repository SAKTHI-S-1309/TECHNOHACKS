# FILE ORGANIZER

## Problem Statement
Develop a Python script that organizes files in a specified directory based on their file types.



```python
import os
import shutil

path = input("Enter folder path: ")

file_types = {
    "Images": [".jpg", ".jpeg", ".png", ".gif"],
    "Documents": [".pdf", ".docx", ".txt"],
    "Videos": [".mp4", ".mkv"],
    "Audio": [".mp3", ".wav"]
}

# Organize files
for file in os.listdir(path):

    file_path = os.path.join(path, file)

    # Skip folders
    if os.path.isdir(file_path):
        continue

    extension = os.path.splitext(file)[1].lower()

    moved = False

    for folder, extensions in file_types.items():

        if extension in extensions:

            folder_path = os.path.join(path, folder)

            # Create folder if not exists
            os.makedirs(folder_path, exist_ok=True)

            shutil.move(file_path, os.path.join(folder_path, file))

            print(f"{file} moved to {folder}")
            moved = True
            break

    if not moved:
        others_path = os.path.join(path, "Others")
        os.makedirs(others_path, exist_ok=True)

        shutil.move(file_path, os.path.join(others_path, file))

        print(f"{file} moved to Others")
```

## Conclusion
The File Organizer application works by scanning all files in a user-specified directory using the os library. Based on the file extension, the program identifies the file type such as images, documents, videos, or audio files. It automatically creates separate folders for each category if they do not already exist and moves the files into their respective folders using the shutil library. Files that do not match predefined categories are moved to an “Others” folder. This project demonstrates file handling and automation using Python.
