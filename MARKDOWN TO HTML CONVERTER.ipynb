# MARKDOWN TO HTML CONVERTER

## Problem Statement
Create a Python script that converts Markdown files into HTML.



```python
def markdown_to_html(text):
    lines = text.split("\n")
    html_output = []

    for line in lines:

        # Heading
        if line.startswith("# "):
            line = "<h1>" + line[2:] + "</h1>"

        elif line.startswith("## "):
            line = "<h2>" + line[3:] + "</h2>"

        elif line.startswith("### "):
            line = "<h3>" + line[4:] + "</h3>"

        # Bold
        while "**" in line:
            start = line.find("**")
            end = line.find("**", start + 2)
            if end == -1:
                break
            bold_text = line[start+2:end]
            line = line[:start] + "<b>" + bold_text + "</b>" + line[end+2:]

        # Italic
        while "*" in line:
            start = line.find("*")
            end = line.find("*", start + 1)
            if end == -1:
                break
            italic_text = line[start+1:end]
            line = line[:start] + "<i>" + italic_text + "</i>" + line[end+1:]

        html_output.append(line)

    return "\n".join(html_output)

print("Enter Markdown (press Enter twice to stop):")

lines = []
while True:
    line = input()
    if line == "":
        break
    lines.append(line)

md_text = "\n".join(lines)

result = markdown_to_html(md_text)

print("\n HTML OUTPUT \n")
print(result)
```

## Conclusion
The program reads multiple lines of Markdown input from the user. Each line is processed separately to detect Markdown syntax like headings (#), bold (**text**), and italic (*text*). Instead of merging everything into one long string, the program preserves line structure using a list. After conversion, each line is joined with proper new lines and displayed neatly in HTML format. This ensures the output is readable and structured instead of appearing in a single line.
