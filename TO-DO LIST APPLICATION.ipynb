# TO-DO LIST APPLICATION

## Problem Statement: 
Build a command-line To-Do list application where users can add, view, and remove tasks.

```python
# Store tasks in a list
tasks = []

# Add tasks
def add_task():
    task = input("Enter task: ")
    tasks.append(task)
    print("Task added successfully!")

# View tasks
def view_tasks():
    if len(tasks) == 0:
        print("No tasks available.")
    else:
        print("\nTo-Do List:")
        for index, task in enumerate(tasks, start=1):
            print(f"{index}. {task}")

# Remove tasks
def remove_task():
    view_tasks()

    if len(tasks) == 0:
        return

    try:
        task_num = int(input("Enter task number to remove: "))

        if 1 <= task_num <= len(tasks):
            removed = tasks.pop(task_num - 1)
            print(f"Task '{removed}' removed successfully!")
        else:
            print("Invalid task number!")

    except ValueError:
        print("Please enter a valid number.")


while True:
    print("\n=== To-Do List Menu ===")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Remove Task")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_task()

    elif choice == "2":
        view_tasks()

    elif choice == "3":
        remove_task()

    elif choice == "4":
        print("Exiting To-Do List Application...")
        break

    else:
        print("Invalid choice! Please try again.")
```

## Conclusion
This project demonstrates a simple command-line To-Do List application using Python. Users can add, view, and remove tasks efficiently.
