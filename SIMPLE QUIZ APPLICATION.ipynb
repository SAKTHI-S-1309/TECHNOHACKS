# SIMPLE QUIZ APPLICATION

## Problem Statement
Create a command-line quiz application that quizzes users on various topics and gives scores.



```python
questions = {
    "What is the capital of India?": "Delhi",
    "Which programming language is known as the language of AI?": "Python",
    "What is 5 + 3?": "8",
    "Who developed Python?": "Guido van Rossum",
    "Which planet is known as the Red Planet?": "Mars"
}

score = 0

print(" QUIZ APPLICATION \n")

for question, answer in questions.items():
    user_answer = input(question + " : ")

    if user_answer.strip().lower() == answer.lower():
        print("Correct Answer!\n")
        score += 1
    else:
        print(f"Wrong Answer! Correct answer is {answer}\n")

print(" QUIZ RESULT ")
print(f"Your Score: {score}/{len(questions)}")

# Feedback
if score == len(questions):
    print("Excellent Performance!")
elif score >= 3:
    print("Good Job!")
else:
    print("Keep Practicing!")
```

## Conclusion
The Quiz Application works through a command-line interface where users answer a series of questions. The program stores questions and answers using a dictionary and displays each question one by one to the user. After receiving the user’s answer, the program compares it with the correct answer and updates the score accordingly. At the end of the quiz, the total score and performance feedback are displayed based on the user’s results. The project demonstrates the use of dictionaries, loops, conditional statements, and user input handling in Python.
